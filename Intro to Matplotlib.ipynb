```python
# Import Heart Disease
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# create a histogram of the age column.
heart_disease["age"].plot(kind="hist", bins=50);
```

```python
heart_disease.dtypes
```

```python
heart_disease.plot(kind="hist", figsize=(5,20), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target", # gives black/white color, 0/1 for the target value.
             figsize=(10,6));
```

```python
# Method 2
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol",c="target",ax=ax)
ax.set_xlim([45,100]);
```

```python
## OO method from scratch.
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# We make a subplot, age vs chol and age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10))

# ax0 - age vs chol

scatter = ax0.scatter(over_50["age"], over_50["chol"], c=over_50["target"])
ax0.set(title="Age vs Cholestrol", xlabel="Age", ylabel="Cholestrol")
ax0.axhline(over_50["chol"].mean(), linestyle='--')
ax0.legend(*scatter.legend_elements(), title="Target")


# ax1 - age vs thalach

scatter = ax1.scatter(over_50["age"], over_50["thalach"], c=over_50["target"])
ax1.set(title="Age vs Thalach", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle='--')

# Add a title to the figure.

fig.suptitle("Heart Disease Analysis");
```
